```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

df.columns = [col.replace(" ", "_").replace("(", "").replace(")", "") for col in df.columns]

model = smf.ols('Amount_Collected ~ Campaign_Email + Campaign_Flyer + Campaign_Phone + Sales_Contact_1 + Sales_Contact_2 + Sales_Contact_3 + Sales_Contact_4 + Sales_Contact_5', data=df).fit()

print(model.summary())
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

model = smf.ols('Amount_Collected ~ Campaign_Email + Campaign_Flyer + Campaign_Phone + Sales_Contact_1 + Sales_Contact_2 + Sales_Contact_3 + Sales_Contact_4 + Sales_Contact_5', data=df).fit()

summary_df = model.summary2().tables[1]

variables_to_keep = summary_df[summary_df['P>|t|'] < 0.5].index

formula = 'Amount_Collected ~ ' + ' + '.join(variables_to_keep[variables_to_keep != 'Intercept'])

new_model = smf.ols(formula, data=df).fit()

print(new_model.summary())
```

## Regression Analysis(Market Sales and Strategies) - Broken for different account types

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd

consolidated_summary = pd.DataFrame()

for acctype in df['Client_Type'].unique():
    temp_data = df[df['Client_Type'] == acctype].copy()
    
    results = smf.ols('Amount_Collected ~ Campaign_Email + Campaign_Flyer + Campaign_Phone + Sales_Contact_1 + Sales_Contact_2 + Sales_Contact_3 + Sales_Contact_4 + Sales_Contact_5', data=temp_data).fit()
    
    summary_df = results.summary2().tables[1]
    
    df1 = summary_df[summary_df['P>|t|'] < 0.05][['Coef.', 'P>|t|']].reset_index()
    
    df1.columns = ['Variable', 'Coefficient (Impact)', 'P>|t|']
    
    df1 = df1[df1['Variable'] != 'Intercept']
    df1 = df1.sort_values('Coefficient (Impact)', ascending=False)
    
    df1['Account Type'] = acctype
    
    consolidated_summary = pd.concat([consolidated_summary, df1], ignore_index=True)

for acctype in consolidated_summary['Account Type'].unique():
    print(f"\n{acctype}\n", "-" * len(acctype))
    print(consolidated_summary[consolidated_summary['Account Type'] == acctype].to_string(index=False))
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd

consolidated_summary = pd.DataFrame()

for acctype in df['Client_Type'].unique():
    print(acctype)
    temp_data = df[df['Client_Type'] == acctype].copy()
    results = smf.ols('Amount_Collected ~ Campaign_Email + Campaign_Flyer + Campaign_Phone + Sales_Contact_1 + Sales_Contact_2 + Sales_Contact_3 + Sales_Contact_4 + Sales_Contact_5', data=temp_data).fit()
    
    summary_df = results.summary2().tables[1]
    df1 = summary_df[summary_df['P>|t|'] < 0.05][['Coef.', 'P>|t|']].reset_index()
    df1.columns = ['Variable', 'Coefficient (Impact)', 'P>|t|']
    df1 = df1[df1['Variable'] != 'Intercept']
    df1 = df1.sort_values('Coefficient (Impact)', ascending=False)
    df1['Account Type'] = acctype
    
    consolidated_summary = pd.concat([consolidated_summary, df1], ignore_index=True)

for acctype in consolidated_summary['Account Type'].unique():
    print(f"\n{acctype}\n", "-" * len(acctype))
    print(consolidated_summary[consolidated_summary['Account Type'] == acctype].to_string(index=False))
```

```python
import matplotlib.ticker as ticker  # Import ticker

consolidated_summary['Return on Investment Numeric'] = consolidated_summary['Return on Investment'].replace('[\$,]', '', regex=True).astype(float)

account_types = consolidated_summary['Account Type'].unique()

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
axes = axes.flatten()

for i, account_type in enumerate(account_types):
    df_account_type = consolidated_summary[consolidated_summary['Account Type'] == account_type]
    
    df_account_type_sorted = df_account_type.sort_values(by='Return on Investment Numeric', ascending=False)
    
    sns.barplot(x='Variable', y='Return on Investment Numeric', data=df_account_type_sorted, ax=axes[i], palette='Greys_r', edgecolor='black')
    axes[i].set_title(f'Return on Investment - {account_type}')
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45, ha='right')
    
    axes[i].yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.1f}'))

    for p in axes[i].patches:
        height = p.get_height()
        axes[i].annotate(f'${height:.1f}', (p.get_x() + p.get_width() / 2., height),
                         ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```
